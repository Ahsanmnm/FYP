```python
import os
import pandas as pd

def merge_csv_files(folders):
    # Create an empty list to store the dataframes
    dataframes = []

    # Iterate over the folders
    for folder in folders:
        folder_path = os.path.join('.', folder)  # Assuming the folders are in the same directory as the script

        # Find the CSV file inside the folder
        csv_files = [file for file in os.listdir(folder_path) if file.endswith('Accelerometer.csv')]

        if len(csv_files) == 1:
            csv_file_path = os.path.join(folder_path, csv_files[0])

            # Read the CSV file into a dataframe
            df = pd.read_csv(csv_file_path, sep='\t',usecols=lambda column: column != 'Time (s)')

            # Remove first and last 100 rows
            df = df.iloc[500:-500]

            # Add the dataframe to the list
            dataframes.append(df)

    # Merge the dataframes
    merged_df = pd.concat(dataframes, axis=0)  # Select all columns using axis=0

    # Reset the index
    merged_df.reset_index(drop=True, inplace=True)

    return merged_df

# Example usage
folders = ['S1_MA', 'S1_MU', 'S1_AH']
merged_df = merge_csv_files(folders)

# Print the merged dataframe
print(merged_df)
```

```python
import pandas as pd
import math

def split_dataframe(df, split_size, label):
    # Create an empty list to store the chunks
    chunks = []
    rounded_value = math.floor(len(df)/split_size)
    p=0

    # Split the dataframe into chunks of the specified size
    for i in range(0,rounded_value):
        chunk = df.iloc[p:p+split_size].copy()  # Create a copy of the chunk
          # Assign the specified label to the chunk
        chunks.append(chunk)
        p=p+split_size
    s = len(chunks)    
    
    return chunks , s


```

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_metrics(confusion_matrix):
    # Calculate true positives, false positives, and false negatives
    TP = np.diag(confusion_matrix)
    FP = np.sum(confusion_matrix, axis=0) - TP
    FN = np.sum(confusion_matrix, axis=1) - TP

    # Calculate precision, recall, and F1 score
    precision = TP / (TP + FP)
    recall = TP / (TP + FN)
    f1_score = 2 * (precision * recall) / (precision + recall)

    # Calculate specificity
    TN = np.sum(confusion_matrix) - (TP + FP + FN)
    specificity = TN / (TN + FP)

    return f1_score, recall, specificity

confusion_matrix = np.array([[12, 0, 0, 0],
                             [0, 7, 1, 0],
                             [0, 0, 14, 0],
                             [0, 0, 0, 10]])

f1_score, recall, specificity = calculate_metrics(confusion_matrix)

# Plot the metrics
labels = ['Class 0', 'Class 1', 'Class 2', 'Class 3']
x = np.arange(len(labels))

plt.figure(figsize=(10, 6))
plt.bar(x - 0.2, f1_score, width=0.2, label='F1 Score')
plt.bar(x, recall, width=0.2, label='Sensitivity')
plt.bar(x + 0.2, specificity, width=0.2, label='Specificity')

plt.xlabel('Classes')
plt.ylabel('Metrics')
plt.title('Metrics for each Class')
plt.xticks(x, labels)
plt.legend()
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

# Siamese network architecture
input_shape = (600, 3)

# Define the siamese base network
base_network = models.Sequential()
base_network.add(layers.LSTM(64, input_shape=input_shape))
base_network.add(layers.Dense(128, activation='relu'))

# Define the siamese input layers
input_a = layers.Input(shape=input_shape)
input_b = layers.Input(shape=input_shape)

# Obtain the feature embeddings from the base network for the inputs
feature_a = base_network(input_a)
feature_b = base_network(input_b)

# Calculate the Euclidean distance between the feature embeddings
distance = tf.reduce_sum(tf.square(feature_a - feature_b), axis=-1, keepdims=True)

# Define the siamese model
siamese_model = models.Model(inputs=[input_a, input_b], outputs=distance)

# Compile the siamese model
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
siamese_model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])

history = model.fit(X_train, y_train, epochs=10)

test_loss, test_accuracy = model.evaluate(X_test, y_test)
print('Test Loss:', test_loss)
print('Test Accuracy:', test_accuracy)


import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Get the predicted probabilities
y_pred_probs = model.predict(X_test)
# Get the predicted labels
y_pred = np.argmax(y_pred_probs, axis=1)

# Create the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()

```
